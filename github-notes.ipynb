refactor: code change that neither fixes a bug nor adds a feature

perf: performance improvement

test: adding or fixing tests

chore: build process or auxiliary tool changes

# Version handling

## Revert to old commit locally

```python
git fetch --all
git reset --hard <SHAsum of your commit>
```

## Checkout to another commit
shal can be found with 'git log'

```python
git checkout <shal>

// creates new branch
git checkout -b <new_branch_name> <sha1>
```

## Return to latest commit

```python
git checkout master
```

## Revert old commit to remote (destroy commits)
NOTE: All history past the old commit will be destroyed

```python
git push --force origin HEAD:main
```

## Revert to old commit to remote (keep commits)
If you didnt create a branch when reverting, otherwise skip creating a branch:

```python
git checkout -b revert-and-edit
git add .
git commit -m "Changes based on earlier commit"

git checkout revert-and-edit
git rebase main
git checkout main
git merge revert-and-edit

git push origin main
```

# Branches
OBS: När du redigerar locally, kom ihåg att commita innan du bytar branch!

## create branch

```python
git branch <new branch>
```

## create branch from a commit

```python
git branch <new-branch-name> <commit-hash>
```

## navigate branches

```python
git checkout <branch>
```

## see all local branches

```python
git branch
```

## see all remote branches

```python
git branch -r
```

## push branch to remote

```python
git push --set-upstream origin <branch to push>
```

## force update local to match remote

```python
git reset --hard HEAD
git pull origin <branch du vill force updatera>
```

## merge local branch

```python
# committa först din färdiga branch
git checkout <master>
git pull origin <master>
git merge test
git push origin <master>
```

## rename branch

```python
# navigera först till branch att byta namn på
git branch -m new-name
```

## Pull remote branch

```python
git fetch --all
git checkout -b <branch name> origin/<branch name>
```

## delete local branch

```python
git branch --delete <branchname>
```

## delete remote branch

```python
git push origin -d <remote branch>
```

# Tracking changes with git

## git status
lists the (tracked) files that have been changed but not yet committed. Also checks if behind, ahead or diverged.

## git diff
prints a diff between the latest commit and the current files in the workspace, or between
two commits. By default git diff shows only unstaged changes. Use git diff --cached
to show already staged changes.

## git checkout
moves to another (previously committed) version of the repository. Can also be
used to revert a file in the working directory to the last commit, discarding any changes.

## git log
Checks local commit log

## git log origin/main
Checks remote commit log

```python
git fetch --all
```
